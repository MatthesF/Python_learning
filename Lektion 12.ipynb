```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = sns.load_dataset("iris")
```

```python
x = df.drop("species",axis=1).values
y = df["species"].values
```

# PCA without scaling

```python
pca = PCA(n_components=4)
PC = pca.fit_transform(x)
```

```python
fig, ax = plt.subplots(1,2,figsize=(12,4))

ax[0].bar(range(0,4),pca.explained_variance_ratio_)
#ax[0].set_xticks(range(4),[f"PC{i+1}" for i in range(4)])
ax[0].set(title="Explained variance ratio")
for i in set(y):
    ax[1].scatter(PC[y==i,0],PC[y==i,1],label=i)

ax[1].set(ylabel="PC1",xlabel="PC0",title="PC of Iris")
ax[1].legend();
```

```python
sum(pca.explained_variance_ratio_[range(2)])
```

# With scaling

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
pca = PCA(n_components=4)
PC = pca.fit_transform(X_scaled)
```

```python
fig, ax = plt.subplots(1,2,figsize=(12,4))

ax[0].bar(range(0,4),pca.explained_variance_ratio_)
#ax[0].set_xticks(range(4),[f"PC{i+1}" for i in range(4)])
ax[0].set(title="Explained variance ratio")
for i in set(y):
    ax[1].scatter(PC[y==i,0],PC[y==i,1],label=i)

ax[1].set(ylabel="PC1",xlabel="PC0",title="PC of Iris")
ax[1].legend();
```
